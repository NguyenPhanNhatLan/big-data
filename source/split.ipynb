```python
import pandas as pd
```

```python
df = pd.read_csv("D:/HK6/DỮ LIỆU LỚN VÀ ỨNG DỤNG/ĐACK/big-data/source/sgp_flat_resale.csv")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df = df.reset_index(drop=True)
df['transaction_id'] = df.index + 1
```

```python
location = (
    df[['town', 'street_name', 'block']]
    .drop_duplicates()
    .reset_index(drop=True)
)

location['location_id'] = location.index + 1

ocation = location[
    ['location_id', 'town', 'street_name', 'block']
]
```

```python
flat = (
    df[['flat_type', 'flat_model', 'storey_range', 'floor_area_sqm']]
    .drop_duplicates()
    .reset_index(drop=True)
)

flat['flat_id'] = flat.index + 1

flat = flat[
    ['flat_id', 'flat_type', 'flat_model', 'storey_range', 'floor_area_sqm']
]
```

```python
lease = (
    df[['lease_commence_date', 'remaining_lease']]
    .drop_duplicates()
    .reset_index(drop=True)
)

lease['lease_id'] = lease.index + 1

lease = lease[
    ['lease_id', 'lease_commence_date', 'remaining_lease']
]
```

```python
time = (
    df[['month']]
    .drop_duplicates()
    .reset_index(drop=True)
)

time['time_id'] =time.index + 1

time['month_date'] = pd.to_datetime(
    time['month']
)

time['year'] = time['month_date'].dt.year
time['month_num'] = time['month_date'].dt.month
time['quarter'] = time['month_date'].dt.quarter

time = time[
    ['time_id', 'month', 'year', 'month_num', 'quarter']
]
```

```python
resale = df.merge(
    location,
    on=['town', 'street_name', 'block'],
    how='left'
)
```

```python
resale = resale.merge(
    flat,
    on=['flat_type', 'flat_model'],
    how='left'
)
```

```python
resale = resale.merge(
    lease,
    on=['lease_commence_date', 'remaining_lease'],
    how='left'
)
```

```python
resale = resale.merge(
    time[['time_id', 'month']],
    on='month',
    how='left'
)
```

```python
resale = resale[
    [
        'transaction_id',
        'time_id',
        'location_id',
        'flat_id',
        'lease_id',
        'resale_price'
    ]
]
```

```python
print(location.shape)
print(flat.shape)
print(lease.shape)
print(time.shape)
print(resale.shape)
```

```python
location.to_csv(
    "location.csv",
    index=False
)

flat.to_csv(
    "flat.csv",
    index=False
)

lease.to_csv(
    "lease.csv",
    index=False
)

time.to_csv(
    "time.csv",
    index=False
)

resale.to_csv(
    "resale.csv",
    index=False
)
```
