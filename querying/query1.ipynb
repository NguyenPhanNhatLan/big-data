```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, count, when, mean, stddev, isnan,
    sum as spark_sum, 
    round as spark_round, 
    min as spark_min, 
    max as spark_max
)

# Khởi tạo một Session hoàn toàn mới chạy trên máy local 
spark = (
    SparkSession.builder
    .appName("Query_Local")
    .master("local[*]")  
    .config("spark.driver.memory", "4g")
    .config("spark.sql.adaptive.enabled", "false") 
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")
print("-> Đã kích hoạt Spark Local thành công!")

path_to_data = "D:/HK6/DỮ LIỆU LỚN VÀ ỨNG DỤNG/ĐACK/big-data/data_source/" 

df_resale = spark.read.csv(path_to_data + "resale.csv", header=True, inferSchema=True)
df_flat = spark.read.csv(path_to_data + "flat.csv", header=True, inferSchema=True)
df_lease = spark.read.csv(path_to_data + "lease.csv", header=True, inferSchema=True)
df_location = spark.read.csv(path_to_data + "location.csv", header=True, inferSchema=True)
df_time = spark.read.csv(path_to_data + "time.csv", header=True, inferSchema=True)

df_resale.createOrReplaceTempView("resale")
df_flat.createOrReplaceTempView("flat")
df_lease.createOrReplaceTempView("lease")
df_location.createOrReplaceTempView("location")
df_time.createOrReplaceTempView("time")

print("Đã load xong dữ liệu và sẵn sàng query!")
```

#### Top 10 town có giá resale trung bình cao nhất

```python
query = spark.sql(""" 
select l.town, round(avg(r.resale_price), 2) as avg_price
from resale r
join location l on r.location_id = l.location_id
group by l.town
order by avg_price desc
limit 10""")
query.show()
```

#### Top 10 town có giá resale trung bình thấp nhất

```python
query = spark.sql(""" 
select l.town, round(avg(r.resale_price), 2) as avg_price
from resale r
join location l on r.location_id = l.location_id
group by l.town
order by avg_price asc
limit 10""")
query.show()
```

#### Top 20 căn hộ có giá resale cao nhất

```python
query = spark.sql(""" 
select * 
from resale r
join location l on r.location_id = l.location_id
join flat f on r.flat_id = f.flat_id
join time t on r.time_id = t.time_id
join lease ls on r.lease_id = ls.lease_id
order by r.resale_price desc
limit 20""")
query.show()
```

#### Top 2 căn hộ đắt tiền nhất mỗi năm

```python
query = spark.sql(""" 
with t1 as
(select r.resale_price, f.flat_type, f.floor_area_sqm, l.town, t.year,
row_number() over (partition by t.year order by r.resale_price desc) as ranking
from resale r
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id
join time t on r.time_id = t.time_id)

select year, town, flat_type, floor_area_sqm, resale_price
from t1 where ranking in (1, 2)
order by year""")
query.show()
```

#### Loại căn hộ phổ biến nhất ở từng town

```python
query = spark.sql(""" 
with t1 as
(select f.flat_type, l.town, 
count(f.flat_id) as flat_count,
row_number() over (partition by l.town order by count(f.flat_id) desc) as rn
from resale r
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id
group by f.flat_type, l.town)

select town, flat_type, flat_count
from t1
where rn = 1
order by town 
""")
query.show()
```

#### Loại tầng phổ biến nhất ở từng town

```python
query = spark.sql("""
with t1 as(
select l.town,
        case
            when f.storey_range in ('01 TO 03', '04 TO 06') then 'Low Floor'
            when f.storey_range in ('07 TO 09', '10 TO 12') then 'Mid Floor'
            else 'High Floor'
        end as floor_group
from resale r
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id),

t2 as (
select town, floor_group, count(*) as count_flat,
row_number() over (partition by town
                  order by count(*) desc) as rn
from t1
group by town, floor_group
)
select town, floor_group, count_flat
from t2
where rn = 1
order by town
                  
""")
query.show()

```

#### Giá trung bình của từng loại căn hộ ở từng town

```python
query = spark.sql(""" 
select l.town, f.flat_type, round(avg(r.resale_price), 2) as avg_price
from resale r
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id
group by l.town, f.flat_type
order by l.town
""")
query.show()
```

#### Con đường đắt giá nhất ở từng town

```python
query = spark.sql(""" 
with t1 as
(select l.town, l.street_name, round(avg(r.resale_price), 2) as avg_price,
row_number() over (partition by l.town order by avg(r.resale_price) desc) as rn
from resale r
join location l on r.location_id = l.location_id
group by l.town, l.street_name
                  )

select street_name, town, avg_price
from t1
where rn = 1
order by avg_price desc
""")
query.show()
```

#### Giá nhà trung bình mỗi năm và % tăng trưởng

```python
query = spark.sql(""" 
with t1 as
(select t.year, round(avg(r.resale_price), 2) as avg_price,
lag(round(avg(r.resale_price), 2)) over (order by t.year) as prev_year_avg_price
from resale r
join time t on r.time_id = t.time_id
group by t.year)

select year, avg_price, prev_year_avg_price,
round(((avg_price - prev_year_avg_price) / prev_year_avg_price) * 100, 2) as pct_change
from t1
""")
query.show()
```

#### Top 2 town có price range lớn nhất theo diện tích nhà



```python
query = spark.sql(""" 
with t1 as
(select 
        CASE
            when f.floor_area_sqm < 70 THEN '< 70 sqm'
            when f.floor_area_sqm < 80 and f.floor_area_sqm >= 70 THEN '70-79 sqm'
            when f.floor_area_sqm < 90 and f.floor_area_sqm >= 80 THEN '80-89 sqm'
            else '>= 90 sqm'
        END AS area_group, 
        min(r.resale_price) as min_price, max(r.resale_price) as max_price
from resale r
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id
group by area_group)

select area_group, max_price - min_price as price_range
from t1
order by price_range desc           
""")
query.show()
```

```python
query = spark.sql("""
SELECT
    CASE
        when f.floor_area_sqm < 70 THEN '< 70 sqm'
        when f.floor_area_sqm < 80 and f.floor_area_sqm >= 70 THEN '70-79 sqm'
        when f.floor_area_sqm < 90 and f.floor_area_sqm >= 80 THEN '80-89 sqm'
        else '>= 90 sqm'
    END AS area_group,

    COUNT(*) AS num_transactions

FROM resale r
JOIN flat f
    ON r.flat_id = f.flat_id

GROUP BY
    CASE
        when f.floor_area_sqm < 70 THEN '< 70 sqm'
        when f.floor_area_sqm < 80 and f.floor_area_sqm >= 70 THEN '70-79 sqm'
        when f.floor_area_sqm < 90 and f.floor_area_sqm >= 80 THEN '80-89 sqm'
        else '>= 90 sqm'
    END

ORDER BY num_transactions DESC
""")

query.show()
```

#### Số giao dịch được thực hiện ở mỗi town (thanh khoản)

```python

query = spark.sql("""
select l.town, count(r.transaction_id) as transaction_count
from resale r
join location l on r.location_id = l.location_id
group by l.town 
order by transaction_count desc
""")

query.show()
```

#### Giá/sqm trung bình ở từng town

```python

query = spark.sql("""
select l.town, round(avg(r.resale_price/f.floor_area_sqm), 2) as avg_price_per_sqm
from resale r   
join flat f on r.flat_id = f.flat_id
join location l on r.location_id = l.location_id
group by l.town
order by avg_price_per_sqm desc
""")

query.show()
```

#### Xu hướng lựa chọn loại căn hộ theo thời gian

```python

query = spark.sql("""
with t1 as (
    select t.year, f.flat_type, count(transaction_id) as transaction_count,
    row_number() over (partition by t.year order by count(transaction_id) desc) as rn
    from resale r   
    join flat f on r.flat_id = f.flat_id
    join time t on r.time_id = t.time_id
    group by t.year, f.flat_type
)
select year, flat_type, transaction_count
from t1
where rn = 1
order by year
""")

query.show()
```
